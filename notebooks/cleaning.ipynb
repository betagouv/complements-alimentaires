**Est-ce que le champ `SBSACT_QUANTITE_ARENSEIGNER` est 1 si au moins `SBSACT_QTE_MIN` ou `SBSACT_QTE_MAX` sont spécifiés ?**

Non, il y a 13 substances pour lesquelles ces deux champs sont vides et pourtant la quantité doit être spécifiée => pas transformable en property

Par contre il faudrait sans doute demander la spécification de la quantité lorsque l'un des deux est rempli ?

```python
import pandas as pd
```

```python
substance_df = pd.read_csv("../../csv-data/REF_ICA_SUBSTANCE_ACTIVE.csv")
```

```python
def f(row):
     if not pd.isna(row['SBSACT_QTE_MIN']):
        return 1
     elif not pd.isna(row['SBSACT_QTE_MAX']):
        return 1
     return 0


substance_df['computed_must_specify_qty'] = substance_df[['SBSACT_QTE_MIN', 'SBSACT_QTE_MAX']].apply(f, axis=1)


substance_df[['SBSACT_QUANTITE_ARENSEIGNER', 'computed_must_specify_qty']].sum()
```

```python
# substance_df.query('SBSACT_QUANTITE_ARENSEIGNER' != 'computed_must_specify_qty')
substance_df.loc[(substance_df['SBSACT_QUANTITE_ARENSEIGNER'] != substance_df['computed_must_specify_qty'])][['SBSACT_LIBELLE', 'SBSACT_COMMENTAIRE_PUBLIC', 'computed_must_specify_qty', 'SBSACT_QUANTITE_ARENSEIGNER']]
```
